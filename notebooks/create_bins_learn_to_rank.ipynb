```python
def create_equal_bins_with_gaps(df, column, num_bins, epsilon):
    """
    Create an arbitrary number of bins of equal length within the value range of the specified column,
    with epsilon gaps between the bins.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    column (str): The name of the column to bin.
    num_bins (int): The number of bins to create.
    epsilon (float): The epsilon value to use for the gaps between bin ranges.

    Returns:
    pd.DataFrame: The DataFrame with an additional column for the bins.
    """
    # Determine the range of the data
    min_value = 1
    max_value = 6
    
    bins = np.linspace(1.0,6.0,num_bins+1)
    bins[0]= bins[0]-epsilon
    bins[-1]= bins[-1]+ epsilon
    labels = range(num_bins)

    list_bins = []

    for i in range(num_bins):
        list_bins.append((bins[i]+epsilon,bins[i+1]-epsilon))
    
    df['bins_adm'] = pd.cut(df["adm"], bins=pd.IntervalIndex.from_tuples(list_bins,closed="both"), labels=labels, include_lowest=True)
    df['bins_riv'] = pd.cut(df["riv"], bins=pd.IntervalIndex.from_tuples(list_bins,closed="both"), labels=labels, include_lowest=True)

    #df['bins'] = pd.cut(df[column], bins=pd.IntervalIndex.from_tuples(list_bins,closed="both"), labels=labels, include_lowest=True)

    return df
    

df_bins= create_equal_bins_with_gaps(df,"adm",2,1/4)
```
